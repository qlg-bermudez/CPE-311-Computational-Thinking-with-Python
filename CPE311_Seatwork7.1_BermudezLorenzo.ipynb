Using the meteorite data from the Meteorite_Landings.csv file, update the year column to only contain the year, convert it to a numeric data type, and create a new column indicating whether the meteorite was observed falling before 1970. Set the index to the id column and extract all the rows with IDs between 10,036 and 10,040 (inclusive) with loc <br><br>

- Hint 1: Use year.str.slice() to grab a substring. <br>
- Hint 2: Make sure to sort the index before using loc[] to select the range. <br>

Bonus: There's a data entry error in the year column. Can you find it? (Don't spend too much time on this )


```python
import pandas as pd

filePath = 'C:\\Users\\Enzo\\Downloads\\Browser Downloads\Meteorite_Landings.csv'

df = pd.read_csv(filePath)

# Update the year column to only contain the year (numeric); the year is at index 6-10
df['year'] = df['year'].astype(str).str.slice(6, 10)
df['year'] = pd.to_numeric(df['year'], errors='coerce')

# Created me a new column indicating if the meteorite was observed falling before 1970
# Fell indicates it was observed falling
# Found indicates it was discovered later.
df['observed_fell_before_1970'] = (df['fall'] == 'Fell') & (df['year'] < 1970)

# Set index to the id column
df.set_index('id', inplace=True)
df.sort_index(inplace=True)

# extract rows 
extracted_rows = df.loc[10036:10040]

extracted_rows
```
