```python
PegA = [100, 75, 50, 25] # OG Peg
PegB = [] # Temporary Peg
PegC = [] # Destination Peg
```

```python
def hanoi(n, source, auxiliary, target, source_name, auxiliary_name, target_name):
    if n == 1:
        # Base case: move the single disk from source to target
        move_disk(source, target, source_name, target_name)
    else:
        # Step 1: Move the last disks from PegA to PegB, using PegC as the temporary
        hanoi(n-1, source, target, auxiliary, source_name, target_name, auxiliary_name)

        # Step 2: Move the current largest disk from PegA to PegC
        move_disk(source, target, source_name, target_name)

        # Step 3: Move the last disks from PegB to PegC, using PegA as temporary
        hanoi(n-1, auxiliary, source, target, auxiliary_name, source_name, target_name)
```

```python
# Main Tower of Hanoi function
def move_disk(source, target, source_name, target_name):
    disk = source.pop()
    target.append(disk)
    print(f"Move disk {disk} from {source_name} to {target_name}")
    return disk

# Call the Tower of Hanoi function to move the disks from PegA to PegC, using PegB as auxiliary
hanoi(4, PegA, PegB, PegC, "A", "B", "C")
```

```python
print("\nFinal state of pegs after Tower of Hanoi:")
print(f"PegA: {PegA}")
print(f"PegB: {PegB}")
print(f"PegC: {PegC}")
```
